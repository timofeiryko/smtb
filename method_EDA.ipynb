### Ligands

```python
def extract_ligand(s):
    right_part = s.split('.')[1]
    return right_part.split('_')[-1]
```

```python
subset_df['ligand'] = subset_df.file_name.apply(extract_ligand)
```

# Explore new associations

```python
subset_new_df = subset_df.query('human_disease not in @old_disease_names')
```
